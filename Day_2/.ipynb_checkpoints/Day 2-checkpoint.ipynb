```python
def C_to_F_temp(C): #define used for function (an initializer)
                    #the (C) creates a variable for the function that the user places in the paranthesis
    #F = (C*(9/5)+32)
    fahrenheit = (C*(9/5)+32)
    print(fahrenheit) #gives us the answer after the function runs but may not allow us to reuse
    return fahrenheit #recalls the final fahrenheit, allows you to manipulate/use
```

```python
C_to_F_temp(100)
```

```python
def get_urban(dF,threshold):
    urban= dF.loc[dF['Population_Density'] >= threshold, :]
    return (urban)
```
